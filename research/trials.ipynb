```python
# about ensure annotation
def getProduct(x:int,y:int)->int:
    return x*y

print(getProduct(2,3)) 
getProduct(2,"3") # see here the function accept the int only but even you pass string it makes calculations and give me the results
# such small bug can be findout by using ensure anotation
```

```python
# about ensure annotation
from ensure import ensure_annotations

@ensure_annotations
def getProduct(x:int,y:int)->int:
    return x*y

print(getProduct(2,3)) # No error till here
print(getProduct(2,'3')) 

```
